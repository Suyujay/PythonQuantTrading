```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))
plt.plot(x, y)
plt.title("Sigmoid Activation Function", fontsize=20)
plt.xlabel("Input", fontsize=20)
plt.ylabel("Output", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.tanh(x)
plt.plot(x, y)
plt.title("Tanh Activation Function", fontsize=20)
plt.xlabel("Input", fontsize=20)
plt.ylabel("Output", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.maximum(0, x)
plt.plot(x, y)
plt.title("ReLU Activation Function", fontsize=20)
plt.xlabel("Input", fontsize=20)
plt.ylabel("Output", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import torch
import torch.nn as nn

# 創建一個隨機的輸入張量
input_tensor = torch.tensor([-1, 1, 2])
print(f"input_tensor: {input_tensor}")
# 定義 Sigmoid 激活函數
sigmoid = nn.Sigmoid()
# 使用 Sigmoid 激活函數進行轉換
output_sigmoid = sigmoid(input_tensor)
print(f"output_sigmoid: {output_sigmoid}")
# 定義 ReLU 激活函數
relu = nn.ReLU()
# 使用 ReLU 激活函數進行轉換
output_relu = relu(input_tensor)
print(f"output_relu: {output_relu}")
# 定義 Tanh 激活函數
tanh = nn.Tanh()
# 使用 Tanh 激活函數進行轉換
output_tanh = tanh(input_tensor)
print(f"output_tanh: {output_tanh}")
```
