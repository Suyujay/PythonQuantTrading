```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = 1 / (1 + np.exp(-x))
plt.plot(x, y)
plt.title("Sigmoid Activation Function", fontsize=20)
plt.xlabel("Input", fontsize=20)
plt.ylabel("Output", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.tanh(x)
plt.plot(x, y)
plt.title("Tanh Activation Function", fontsize=20)
plt.xlabel("Input", fontsize=20)
plt.ylabel("Output", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.maximum(0, x)
plt.plot(x, y)
plt.title("ReLU Activation Function", fontsize=20)
plt.xlabel("Input", fontsize=20)
plt.ylabel("Output", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
import torch
import torch.nn as nn

# 
input_tensor = torch.tensor([-1, 1, 2])
print(f"input_tensor: {input_tensor}")
#  Sigmoid 
sigmoid = nn.Sigmoid()
#  Sigmoid 
output_sigmoid = sigmoid(input_tensor)
print(f"output_sigmoid: {output_sigmoid}")
#  ReLU 
relu = nn.ReLU()
#  ReLU 
output_relu = relu(input_tensor)
print(f"output_relu: {output_relu}")
#  Tanh 
tanh = nn.Tanh()
#  Tanh 
output_tanh = tanh(input_tensor)
print(f"output_tanh: {output_tanh}")
```
